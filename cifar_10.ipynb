```
import numpy as np
import tensorflow as tf
import keras
from keras import layers
import matplotlib.pyplot as plt

cifar_10=keras.datasets.cifar10
(X_train,Y_train),(X_test,Y_test)=cifar_10.load_data()

x_train=X_train/255
x_test=X_test/255

print(x_train.shape)
print(x_test.shape)
```

```
import tensorflow as tf
model=tf.keras.models.Sequential([layers.Conv2D(32,(3,3),activation='relu',input_shape=(32,32,3)),layers.MaxPool2D(),layers.Conv2D(32,(3,3),activation='relu'),layers.MaxPool2D(),layers.Conv2D(32,(5,5),activation='relu'),layers.Flatten(),layers.Dense(48,activation='relu'),layers.Dense(10,activation='softmax')])

model.summary()
model.compile(optimizer=tf.keras.optimizers.SGD(),loss=tf.keras.losses.sparse_categorical_crossentropy,metrics=['accuracy'])

history=model.fit(x_train,Y_train,epochs=50,validation_data=(x_test,Y_test))


print(x_train.shape)



```

```
model.evaluate(X_test,Y_test)
```

```
print(history.history['accuracy'])
```

```

plt.plot(history.history['loss'], label='loss')
#plt.plot(history.history['loss'], label = 'loss')
plt.xlabel('Epoch')
plt.ylabel('loss')
plt.title('By mnist_test_data')
plt.legend(loc='upper right')

plt.plot(history.history['accuracy'], label='accuracy')
#plt.plot(history.history['accuracy'], label = 'accuracy')
plt.xlabel('Epoch')
plt.ylabel('accuracy')
plt.title('cifar_10')
plt.legend(loc='upper right')
plt.show()
```
